```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
```

```python
model = load_model('/content/drive/MyDrive/Face Recognisation System/face_recognition_model (1).h5')
```

```python
def preprocess_image(img):
    img = cv2.resize(img, (100, 100))
    img_array = np.expand_dims(img, axis=0)
    return img_array / 255.0
```

```python
from google.colab import files
uploaded = files.upload()

for filename in uploaded.keys():
    print(f'User uploaded file "{filename}" with length {len(uploaded[filename])} bytes')

image_path = next(iter(uploaded.keys()))
img = cv2.imread(image_path)
img_array = preprocess_image(img)

predictions = model.predict(img_array)
predicted_class = np.argmax(predictions)
class_names = ['Ben Afflek', 'Elton Jhon', 'Jerry Seinfeld', 'Madonna', 'Mindy Kaling']
predicted_name = class_names[predicted_class]

print(f'The uploaded image is predicted to be: {predicted_name}')
```
